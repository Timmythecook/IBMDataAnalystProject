### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace this with your actual data)
data = {
    'Age': [23, 30, 35, 40, 45, 50, 55, 60],
    'SOPartFreq': [1, 3, 2, 5, 4, 3, 2, 1],
    'JobSat': [3, 4, 5, 2, 3, 4, 3, 5]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create the bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['SOPartFreq'], s=df['JobSat'] * 100, alpha=0.5, c=df['JobSat'], cmap='viridis')

# Add labels and title
plt.title('Bubble Plot: Age vs. Frequency of Participation (Size = Job Satisfaction)')
plt.xlabel('Age')
plt.ylabel('Frequency of Participation')
plt.colorbar(label='Job Satisfaction')

# Show the plot
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace this with your actual data)
data = {
    'ConvertedCompYearly': [50000, 75000, 100000, 120000, 150000, 200000],
    'JobSat': [3, 4, 5, 2, 4, 3],
    'Age': [23, 30, 35, 40, 45, 50]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create the bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(df['ConvertedCompYearly'], df['JobSat'], s=df['Age'] * 50, alpha=0.5, c=df['Age'], cmap='plasma')

# Add labels and title
plt.title('Bubble Plot: Compensation vs. Job Satisfaction (Size = Age)')
plt.xlabel('Yearly Compensation (ConvertedCompYearly)')
plt.ylabel('Job Satisfaction')
plt.colorbar(label='Age')

# Show the plot
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
data1 = {
    'Age': [23, 30, 35, 40, 45, 50, 55, 60],
    'LanguageHaveWorkedWith': ['Python', 'JavaScript', 'Java', 'C++', 'Ruby', 'Python', 'Java', 'Python'],
    'Frequency': [100, 50, 75, 40, 30, 120, 85, 60]
}

# Create a DataFrame
df1 = pd.DataFrame(data1)

# Calculate the unique programming languages for each age group and their frequencies
languages = df1.groupby(['Age', 'LanguageHaveWorkedWith']).agg({'Frequency': 'sum'}).reset_index()

# Create the bubble plot
plt.figure(figsize=(10, 6))
for language in languages['LanguageHaveWorkedWith'].unique():
    subset = languages[languages['LanguageHaveWorkedWith'] == language]
    plt.scatter(subset['Age'], subset['Frequency'], s=subset['Frequency'] * 10, alpha=0.5, label=language)

# Add labels and title
plt.title('Bubble Plot: Technology Preferences by Age')
plt.xlabel('Age')
plt.ylabel('Frequency of Language Usage')
plt.legend(title='Programming Languages')

# Show the plot
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
# Sample data (replace with your actual data)
data2 = {
    'DatabaseWantToWorkWith': ['MySQL', 'PostgreSQL', 'MongoDB', 'SQLite', 'MySQL', 'MongoDB', 'PostgreSQL'],
    'JobSat': [3, 4, 5, 2, 3, 4, 5],
    'RespondentCount': [50, 60, 70, 40, 55, 80, 90]
}

# Create a DataFrame
df2 = pd.DataFrame(data2)

# Calculate the unique databases and the count of respondents for each
database_counts = df2.groupby(['DatabaseWantToWorkWith', 'JobSat']).agg({'RespondentCount': 'sum'}).reset_index()

# Create the bubble plot
plt.figure(figsize=(10, 6))
for db in database_counts['DatabaseWantToWorkWith'].unique():
    subset = database_counts[database_counts['DatabaseWantToWorkWith'] == db]
    plt.scatter(subset['JobSat'], subset['RespondentCount'], s=subset['RespondentCount'] * 10, alpha=0.5, label=db)

# Add labels and title
plt.title('Bubble Plot: Preferred Databases vs. Job Satisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Number of Respondents')
plt.legend(title='Databases')

# Show the plot
plt.show()
```
