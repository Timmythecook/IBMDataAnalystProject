<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv")

# Drop missing values for ConvertedCompYearly
df = df.dropna(subset=["ConvertedCompYearly"])

# Set figure size
plt.figure(figsize=(10, 6))

# Plot histogram with KDE
sns.histplot(df["ConvertedCompYearly"], bins=50, kde=True, color="skyblue", log_scale=True)

# Formatting
plt.xlabel("Yearly Compensation (log scale)")
plt.ylabel("Number of Respondents")
plt.title("Distribution of Yearly Compensation")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
import pandas as pd

# Load the dataset
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv")

# Drop missing values
df = df.dropna(subset=["Employment", "ConvertedCompYearly"])

# Filter only full-time employees
full_time_df = df[df["Employment"] == "Employed full-time"]

# Calculate median compensation
median_compensation = full_time_df["ConvertedCompYearly"].median()
print(f"Median Compensation for Full-Time Employees: {median_compensation:.2f}")
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Drop missing values for country and compensation
df = df.dropna(subset=["Country", "ConvertedCompYearly"])

# Select top 10 countries by respondent count for readability
top_countries = df["Country"].value_counts().index[:10]
df_filtered = df[df["Country"].isin(top_countries)]

# Box plot to compare salary distribution by country
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_filtered, x="Country", y="ConvertedCompYearly", showfliers=False)
plt.xticks(rotation=45, ha="right")
plt.xlabel("Country")
plt.ylabel("Yearly Compensation")
plt.title("Compensation Distribution by Country")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
