
### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from the provided URL
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Map Age categories to numeric values
age_map = {
    "18-24 years old": 21,
    "25-34 years old": 29,
    "35-44 years old": 39,
    "45-54 years old": 49,
    "55-64 years old": 59,
    "65 years or older": 70,
    "Under 18 years old": 16,
    "Prefer not to say": None  # Handle missing data
}

# Create a new column 'AgeNumeric' for numeric age values
df['AgeNumeric'] = df['Age'].map(age_map)

# Drop rows with NaN values in 'AgeNumeric' or 'JobSatPoints_6'
df = df.dropna(subset=['AgeNumeric', 'JobSatPoints_6'])

# Scatter plot for Age vs Job Satisfaction
plt.figure(figsize=(18, 10))  # Adjusting figure size for better visualization

# Create the scatter plot
sns.scatterplot(x='AgeNumeric', y='JobSatPoints_6', data=df, s=100, edgecolor='w', hue='JobSatPoints_6', palette='viridis')

# Title and labels
plt.title('Scatter Plot: Age vs. Job Satisfaction', fontsize=20)
plt.xlabel('Age', fontsize=15)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=15)

# Show the plot
plt.show()
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from the provided URL
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Map 'Age' into numeric values (same as before for 'AgeNumeric')
age_map = {
    "18-24 years old": 21,
    "25-34 years old": 29,
    "35-44 years old": 39,
    "45-54 years old": 49,
    "55-64 years old": 59,
    "65 years or older": 70,
    "Under 18 years old": 16,
    "Prefer not to say": None
}
df['AgeNumeric'] = df['Age'].map(age_map)

# Drop rows with NaN values in 'AgeNumeric' or 'JobSatPoints_6'
df = df.dropna(subset=['AgeNumeric', 'JobSatPoints_6'])

# Scatter plot with regression line
plt.figure(figsize=(18, 10))  # Adjusting the figure size for better visualization

# Scatter plot with trend line using seaborn
sns.regplot(x='AgeNumeric', y='JobSatPoints_6', data=df, scatter_kws={'s': 100, 'edgecolor': 'w'}, line_kws={'color': 'red'})

# Title and labels
plt.title('Scatter Plot with Trend Line: Age vs Job Satisfaction', fontsize=20)
plt.xlabel('Age', fontsize=15)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=15)

# Show the plot
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the provided URL
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Convert 'ConvertedCompYearly' to numeric, coercing errors to NaN
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

# Convert 'JobSatPoints_6' to numeric, coercing errors to NaN
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Handle 'Age' column which contains age ranges (e.g., '35-44 years old')
# We will map these categories to the midpoint of the range
age_mapping = {
    'Under 18 years old': 17, 
    '18-24 years old': 21, 
    '25-34 years old': 29,
    '35-44 years old': 39, 
    '45-54 years old': 49, 
    '55-64 years old': 59, 
    '65 years or older': 70
}
df['Age'] = df['Age'].map(age_mapping)

# Drop rows where any of the relevant columns are NaN
df_cleaned = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'Age'])

# Check if there are any rows left after cleaning
print(f"Data after cleaning: {df_cleaned.shape[0]} rows")

# Check the cleaned data to ensure it looks correct
print(df_cleaned[['ConvertedCompYearly', 'JobSatPoints_6', 'Age']].head())

# Bubble plot of Compensation vs. Job Satisfaction with Age as bubble size
plt.figure(figsize=(12, 8))
plt.scatter(x=df_cleaned['ConvertedCompYearly'], 
            y=df_cleaned['JobSatPoints_6'], 
            s=df_cleaned['Age'] * 10,  # Bubble size based on Age (scaled for better visibility)
            alpha=0.5, 
            c=df_cleaned['JobSatPoints_6'], cmap='viridis',  # Color by job satisfaction level
            edgecolors='w', linewidth=0.5)

# Title and labels
plt.title('Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size', fontsize=16)
plt.xlabel('Yearly Compensation (ConvertedCompYearly)', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.colorbar(label='Job Satisfaction')  # Add color bar to show job satisfaction scale
plt.show()

```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the provided URL
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Convert 'JobSatPoints_6' to numeric, coercing errors to NaN
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Drop rows with missing job satisfaction data
df_cleaned = df.dropna(subset=['JobSatPoints_6'])

# Extracting programming languages data
# Split the languages in 'LanguageHaveWorkedWith' by commas
language_data = df_cleaned['LanguageHaveWorkedWith'].dropna().str.split(',')

# Flatten the list of languages into a DataFrame
language_expanded = language_data.explode().reset_index(drop=True)

# Assign corresponding job satisfaction scores to each language
job_satisfaction = df_cleaned.loc[language_data.index, 'JobSatPoints_6'].reset_index(drop=True)

# Combine into a DataFrame
language_df = pd.DataFrame({'Language': language_expanded, 'JobSatPoints_6': job_satisfaction})

# Group by language to get the count of people using each language and their average job satisfaction score
language_summary = language_df.groupby('Language').agg(
    count=('JobSatPoints_6', 'size'),
    avg_satisfaction=('JobSatPoints_6', 'mean')
).reset_index()

# Sorting the data by count to make the plot more readable
language_summary = language_summary.sort_values('count', ascending=False)

# Scatter plot for programming languages by job satisfaction
plt.figure(figsize=(14, 8))
plt.scatter(language_summary['count'], language_summary['avg_satisfaction'], 
            alpha=0.7, c=language_summary['avg_satisfaction'], cmap='viridis', s=100, edgecolors='w')

# Title and labels
plt.title('Popularity of Programming Languages vs Job Satisfaction', fontsize=16)
plt.xlabel('Number of People Using Language', fontsize=12)
plt.ylabel('Average Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.xticks(rotation=90)
plt.colorbar(label='Job Satisfaction')  # Color bar to represent job satisfaction levels
plt.show()
```
