### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and 'DatabaseWantToWorkWith' contains the survey data
top_databases = df['DatabaseWantToWorkWith'].value_counts().head(5)

# Plotting the pie chart
plt.figure(figsize=(8, 6))
top_databases.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title('Top 5 Databases Respondents Want to Work With')
plt.ylabel('')
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
# Assuming 'DevType' contains the developer types
top_dev_roles = df['DevType'].value_counts().head(5)

# Plotting the pie chart
plt.figure(figsize=(8, 6))
top_dev_roles.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff6666','#ffcc66','#99ff66','#66b3ff','#ff66b3'])
plt.title('Top 5 Developer Roles')
plt.ylabel('')
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
# Assuming 'OpSysProfessionalUse' contains the operating systems used by developers professionally
top_opsys = df['OpSysProfessionalUse'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 6))
top_opsys.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ffb3e6','#ffb366','#b3ffb3','#ff6666','#c2c2f0'])
plt.title('Operating Systems Used by Respondents Professionally')
plt.ylabel('')
plt.show()
```
